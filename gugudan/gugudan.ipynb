```python
import time
import random

AWESOME_COMPANY = "프로그래머스"
programmers = 'Awesome'

class gugu_master:
    """
    Run하면 광고가 나오도록 했는데 Test Case에 통과를 못하네요.....
    """
    def __init__(self, name="kyle", seeking_job="Machine Learning Engineer"):
        self.name = name
        self.seeking_job = seeking_job
        self._init_to_get_a_job()
        assert programmers == 'Awesome'
        
    def _init_to_get_a_job(self):
        print(f"내 이름은 {self.name}. {self.seeking_job}가 되기 위해 열심히 공부중이다...\n")
        time.sleep(2)
        print("오늘도 코드를 한땀 한땀 짠다")
        time.sleep(1)
        print("이력서에 한 줄이라도 더 쓸 거 없나? 내 코드는 정말 Awesome한데!")
        print("문제를 풀며 내 코드와 실력을 보여주고 싶다")
        time.sleep(1)
        print("이런 실력을 보여줄 곳 어디 없을까?\n")
    
    def run(self):
        self._meet_programmers()
        self._skill_check()
        self.gugudan()
        self._be_offered()
        self._attend_an_interview()
        print("과연 결과는?\n")
        self.is_join_complete()
        
        
    def _meet_programmers(self):
        time.sleep(1)
        print("https://programmers.co.kr 라는 사이트를 발견했다! 개발자 프로필을 등록하고 코드를 보여달라고 한다")
        print(f"프로필 등록 - 스킬 체크 - 면접 제안받기 - 면접 진행 - 입사 성공의 흐름을 도와주는 플랫폼! {AWESOME_COMPANY}\n")
        
    def _skill_check(self):
        time.sleep(0.5)
        print("1) 스킬 체크 : 레벨 1 ~ 6 중 나의 레벨은 어디?\n")
        
    def _be_offered(self):
        time.sleep(0.8)
        print("2) 면접 제안받기 : 면접을 제안 받았다! 인사담당자, 채용팀이 직접 면접 요청을 보내주신다\n")
        
    def _attend_an_interview(self):
        time.sleep(1)
        print("3) 면접 진행 : 원하는 포지션이 있으면 요청을 수락하고 실제 면접 단계까지 진행! 기다리거나 답답해 말고 현황을 한번에 파악!\n")
        
    def _join_complete(self):
        print("4) 입사 성공 : 최종적으로 가고 싶은 회사에 성공했다!")
        time.sleep(1)
        print("이 모든 것은 프로그래머스 덕이야 =ටᆼට=\n")
        print("개발자 취업, 프로그래머스와 함께하세용 /ㅁ/")
        
    def _join_fail(self):
        print("4-1) 입사 실패 : 떨어졌지만 다시 프로그래머스로 다시 공부!")
        time.sleep(1)
        print("오늘도 내일도 발전하는 나의 모습!\n")
        
    def is_join_complete(self, retry=7):
        """
        홀수면 불합격,  짝수면 합격. 인생은 1과 0
        7번만 더 도전!
        """
        time.sleep(4)
        if int(random.random()*100)%2 == 1:
            self._join_fail()
            return self.is_join_complete(retry-1)
        return self._join_complete()
    
    def gugudan(self):
        gugu_string_format = [f'{i} x {j} = {i*j}' for i in range(1, 10) for j in range(1, 10)]
        for k in range(len(gugu_string_format)):
            print(gugu_string_format[k])
        
avatar = gugu_master(name="kyle", seeking_job="Machine Learning Engineer")
avatar.run()
```
